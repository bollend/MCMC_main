```python
import numpy as np
import matplotlib.pylab as plt

exp_velocity = np.arange(-10,10,.5)
jet_angle = 60
number_of_gridpoints = 100
polar_angle_gridpoints = np.arange(0,60,1.5)
power = 2
jet_cavity_angle=0
velocity_max = 800
velocity_edge =8

for expon in exp_velocity:
    
    pol_velocity = np.zeros(number_of_gridpoints)

    e_exponent = np.exp(np.exp(expon))

    exp_angles = np.abs(polar_angle_gridpoints / jet_angle)**power
    factor = ( e_exponent**-(exp_angles) - e_exponent**-1 )\
                 / ( 1 - e_exponent**-1 )

    pol_velocity = velocity_edge + (velocity_max - velocity_edge) * factor
    plt.plot(polar_angle_gridpoints, pol_velocity)
    
plt.show()

def poloidal_velocity(number_of_gridpoints, power):
    """
    The jet has a single velocity law, with a varying velocity profile
    The velocity is not calculated for the gridpoints in the jet with
    a polar angle smaller than the cavity angle.
    """
    pol_velocity = np.zeros(number_of_gridpoints)
    indices           = np.where(polar_angle_gridpoints > jet_cavity_angle)
    e_exponent          = np.exp(exp_velocity)

    exp_angles = np.abs(polar_angle_gridpoints[indices] / jet_angle)**power

    factor = ( e_exponent**-(exp_angles) - e_exponent**-1 )\
             / ( 1 - e_exponent**-1 )

    pol_velocity[indices] \
            = velocity_edge + (velocity_max - velocity_edge) * factor

    return pol_velocity
```
